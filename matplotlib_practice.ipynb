```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
# Added after to show difference
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# skill Drill
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2)
ax.grid()
ax.legend('Boston')
ax.set_ylim(0, 45)
ax.set_title("PyBer Fare By Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
```

```python
# Scatter Plot (Marlab Method)
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Make bigger (Iterate through y values *3)
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Make bigger refactored (List Comprehension)
# Format (new_list = [expression for item in list if conditional])
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter Plot (Object Oriented)
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Skill Drill

fig, ax = plt.subplots()
ax.set_title('Pyber Fare By Month') 
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)') 
ax.set_xlim(0,50)
plt.gca().invert_yaxis() 
ax.scatter(y_axis, x_axis, edgecolors='black', facecolors='skyblue',  label='Boston', linewidth=2, s = [i * 5 for i in y_axis])
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')


plt.show()
```

```python
# Skill Drill
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.3, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%', shadow=True, counterclock=False, startangle=90)

plt.show()
```
